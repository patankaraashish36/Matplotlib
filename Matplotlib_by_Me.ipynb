# Scatter Plot 

```python
from matplotlib import pyplot as plt
import numpy as np
# import sys
a=np.random.rand(10) # This is X axis 
b=np.arange(1,11)  # This is Y axis 
c=np.random.rand(10)
# print(a)
sizes=[100,1000,345,655,244,100,1000,345,655,244]
colors=[1,2,4,5,7,6,8,3,5,4]
plt.scatter(a,b,s=sizes,c=colors,cmap="Accent",edgecolors="r",linewidths=1,marker="o",alpha=0.7)
plt.scatter(c,b,s=sizes,c=colors,cmap="Blues",edgecolors="g",linewidths=1,marker="o",alpha=0.7)
plt.title("Adore",fontsize=15)
plt.xlabel("A",fontsize=15)
plt.ylabel("P",fontsize=15)
colorBar=plt.colorbar()
colorBar.set_label("Color Meter",fontsize=15)
plt.show()
# print(sys.getsizeof(a
```

# Bar plot

```python
x=["Python","Java",'JavaScript','HTML','CSS','C','C++','R']
y=[90,70,80,100,75,40,60,65]
z=[30,50,20,48,19,37,45,35]
width=0.4
zz=[j+width for j in range(len(x)) ]
color=["y","g","y","b","r","b","r",'g']
l="Programming Languages"
plt.bar(x,y,width,color=color,label=l,edgecolor="black",alpha=1,linewidth=2,linestyle="-",align="center")
plt.bar(zz,z,width,color="r",edgecolor="black",alpha=1,linewidth=2,linestyle="-",align="center")
plt.legend()
plt.xticks(rotation=50)
plt.title("Popularity",fontsize=20)
plt.xlabel("Languages",fontsize=20)
plt.ylabel("Percentage",fontsize=20)

plt.show()
```

```python
a=["Aashish","Golu","Ram","Chetan","Ashwin","Aayush","Piyush"]
a_Age=[21,22,18,17,18,16,18]
a_Hgt=[6,5.5,5,5.5,6.2,6.2,5.5]
width=0.4
d=[j+0.4 for j in range(len(a))]
plt.bar(a,a_Age,width,label="Age",edgecolor="black",linewidth=2)
plt.bar(d,a_Hgt,width,label="Hight",edgecolor="black",linewidth=2)
plt.legend()
plt.title("Friends_Data",fontsize=25,color="g")
plt.xlabel("Names",fontsize=25,color="b")
plt.ylabel("Numbers",fontsize=25,color="b")
plt.xticks(rotation=45)
```
