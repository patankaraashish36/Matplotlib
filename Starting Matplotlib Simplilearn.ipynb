```python
import numpy as np
import pandas as pd
from matplotlib import pylab
```

```python
print(pylab.__version__)
```

```python
"""use numpy to generate a random array"""
a=np.linspace(0,10,25)
print(a)
a.shape
```

```python
b=a*a+2
print(b)
c=np.array([a,b]).reshape(2,25)
print("C =",c)
c.shape
```

```python
'''It only takes 1 argument (b) to draw'''
pylab.plot(a,b,"b")  # b stands for blue or any color
```

```python
'''Drawing the subgraph'''
pylab.subplot(1,2,1)  # The content of the brackets represent (rows,columns,indexs)
pylab.plot(a,b,"b*--")  # Third parameter is represent the color and line style
pylab.subplot(1,2,2)
pylab.plot(b,a,"r+-")
# pylab.subplot(2,2,3)
# pylab.plot(b,a,"b--")
```

```python
'''(1,2,1) and (1,2,2) are significant if we want tiled subplot, without which both graphs will in the same plot'''
# For example
pylab.subplot(1,2,1)  # The content of the brackets represent (rows,columns,indexs)
pylab.plot(a,b,"b*--")  # Third parameter is represent the color and line style
pylab.subplot(1,2,1)
pylab.plot(b,a,"r+--")
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()

ax = fig.add_axes([0.5,0.1,0.8,0.8]) # Control the left right width height of the canvas (from 0 to 1)

ax.plot(a,b,"b*-")
```

```python
# again we can drow subgraph
fig,axes=plt.subplots(nrows=1,ncols=2)  # submap is of 1 row and 2 coulmns
for ax in axes:
    ax.plot(a,b,"b")
```
