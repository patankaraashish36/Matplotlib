```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(50)
y=np.random.rand(50)
# print(y)
title="Scatter Plot"
plt.scatter(x,y)
plt.xlabel("X",fontsize=25)
plt.ylabel("Y",fontsize=25)
plt.title(title,fontsize=25)
plt.show()
```

# Applaying different different color

```python
x=np.arange(7)
y=np.random.rand(7)
# print(y)

colors=["r","y","g","b","r","g","r"]
size=[140,150,160,170,100,160,100]
colors=[10,20,30,40,50,60,70]
# marker parameter is us for the rizing the circules or plots "*" or "+"
# alpha parameter is use for the increse and dicresa the transparance of the plots
# linewidth is use for the increase and decrease the width of the edgecolor

plt.scatter(x,y,c=colors,s=size,cmap="Blues")

l=plt.colorbar( )# how to known the which color is assingn to the which number

l.set_label("ColorBar",fontsize=25)

plt.xlabel("X",fontsize=25)
plt.ylabel("Y",fontsize=25)


title="Scatter Plot"
plt.title(title,fontsize=25)
plt.show()
```

# Pie plot

```python
x=np.arange(10,50,10)
# print(x)
y=['C','C++',"JAVA","PYTHON"]
```

```python
ex=[0.3,0.0,0.0,0.0]
c=["r","b","g","y"]
plt.pie(x,labels=y,colors=c,explode=ex,autopct='%0.2f%%',shadow=True,radius=1,labeldistance=1.1,startangle=30,textprops={'fontsize':13},counterclock=False,wedgeprops={'linewidth':4,'width':1}
       ,rotatelabels=False,)
plt.title('AP',fontsize=15)
plt.legend(loc=2)
plt.show()
```
