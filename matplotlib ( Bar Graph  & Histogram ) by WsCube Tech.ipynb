```python
from matplotlib import pyplot as plt 
import numpy as np
```

# Bar plot

```python
x=["Python","Java","C++","C"]
y=[90,85,70,60]
z=[10,20,30,40]

c=["y","g","r","b"]
width=0.5

plt.xlabel("Languages",fontsize=25)
plt.ylabel("Popularity",fontsize=25)
plt.title("A-P",fontsize=25)
# for graph darkness or color use alpha parameter
# plt.bar(x,y,color=c,width=0.5,align="center",edgecolor="b",linewidth=2,linestyle="--",alpha=0.9)

l="Programming Languages"

plt.bar(x,y,width,color=c,edgecolor="black",label=l)
plt.bar(x,z,width,color="black",label=l)
plt.legend()
plt.show()
```

```python
x=["Python","Java","C++","C"]
y=[90,85,70,60]
z=[10,20,30,40]

width=0.3
p=np.arange(len(x))
w=[j + width for j in p]
# print(w)
c=["y","g","r","b"]

plt.xlabel("Languages",fontsize=25)
plt.ylabel("Popularity",fontsize=25)
plt.title("A-P",fontsize=25)
# for graph darkness or color use alpha parameter
# plt.bar(x,y,color=c,width=0.5,align="edge",edgecolor="b",linewidth=2,linestyle="--",alpha=0.9)
l="Programming Languages"

plt.bar(p,y,color=c,width=width,edgecolor="b",label=l)
plt.bar(w,z,color="black",width=width,label=l)
plt.legend()

plt.xticks(np.arange(4),x,rotation=50)

plt.show()
```

# For Horizontal Bar Graph 

```python
x=["Python","Java","C++","C"]
y=[90,85,70,60]
z=[10,20,30,40]

# width=0.3
# p=np.arange(len(x))
# w=[j + width for j in p]
# print(w)
c=["y","g","r","b"]

plt.xlabel("Languages",fontsize=25)
plt.ylabel("Popularity",fontsize=25)
plt.title("A-P",fontsize=25)
# for graph darkness or color use alpha parameter
# plt.bar(x,y,color=c,width=0.5,align="edge",edgecolor="b",linewidth=2,linestyle="--",alpha=0.9)
l="Programming Languages"

plt.barh(x,y,color=c,edgecolor="b",label=l)
plt.barh(x,z,color="black",label=l)
plt.legend()

plt.xticks(p+0.3/2,x,rotation=50)

plt.show()
```

# Histogram plot

```python
x=np.random.randint(10,60,(50))
print(type(x))
x=list(x)
print(type(x),x)
```

```python
l=[10,20,30,40,50,60]
# bins 
plt.hist(x,bins=l,color="m",edgecolor="r",linewidth=0.4,bottom=10,align="mid",histtype="bar",orientation='vertical',rwidth=0.8,log=False,label="AP",)
plt.axvline(37,color="r",label="line")
plt.legend()
# plt.grid()
plt.title("adore",fontsize=15)
plt.xlabel("A. P.",fontsize=15)
plt.ylabel("P. A.",fontsize=15)
plt.show()
```
